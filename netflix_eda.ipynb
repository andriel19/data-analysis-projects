```python
#Count the number of unique values in the rating column.
df['rating'].nunique()
```

```python
#Filter the dataset to include only rows where type is "Movie"
movies_df = df[df['type'] == "Movie"]
movies_df.head()
```

```python
#Filter the dataset to include only rows where country is "India".
countries_df = df[df['country'] == "India"]
countries_df.head()
```

```python
#Find all shows that were released after the year 2015.
years_df = df[df['release_year'] > 2015]
years_df.head()
```

```python
#Find all shows where the title contains the word "Love"

title_df = df[df['title'] .str.contains("Love")]
title_df.head()
```

```python
# Filter all rows where duration is greater than 90 minutes.
# Keep only movies
movies_df = df[df['type'] == "Movie"].copy()

# Drop rows where duration is NaN
movies_df = movies_df.dropna(subset=['duration'])

# Keep only rows that contain "min"
movies_df = movies_df[movies_df['duration'].str.contains("min")]

# Extract number of minutes and convert to int
movies_df['duration_int'] = movies_df['duration'].str.replace(" min", "", regex=False).astype(int)

# Filter movies longer than 90 minutes
duration_df = movies_df[movies_df['duration_int'] > 90]

duration_df.head()


```

```python
#Drop all rows where the director column is null.
director_df = df.dropna(subset=['director'])
director_df.head()
```

```python
#Fill missing values in the country column with "Unknown".
df['country'] = df['country'].fillna("Unknown")
df.head()
```

```python
#Count how many rows have missing values in the cast column.
df['cast'].isnull().sum()
```

```python
#Replace all missing values in the rating column with "Not Rated".
df['rating'].replace("Not Rated","")
```

```python
#Convert the date_added column to datetime format.

df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce')
```

```python
#Extract the year from the date_added column.
df['year_added'] = df['date_added'].dt.year
```

```python
#Extract the month name from the date_added column.
df['month_added'] = df['date_added'].dt.month_name()
```

```python
df['month_added'] 
```

```python
#Count the number of titles added in the year 2020.
count_2020 = df[df['date_added'].dt.year == 2020]['title'].count()
print(count_2020)
```

```python
most_common_month = df['date_added'].dt.month.value_counts().idxmax()
count = df[df['date_added'].dt.month == most_common_month]
```

```python
# Year-wise distribution
df['year_added'].value_counts().sort_index()
```

```python
# Split multiple countries
df['country'].str.split(',').explode().value_counts().head()
```

```python
# Split genres
df['listed_in'].str.split(',').explode().value_counts().head()
```

```python
# Extract duration
df['duration_min'] = df['duration'].str.extract('(\d+)').astype(float)

# Average movie length
df[df['type']=='Movie']['duration_min'].mean()
```
